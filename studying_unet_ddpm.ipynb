## UNET

의료용 이미지에 주로 많이 사용되는 이미지 세그멘테이션 알고리즘

U-NET: Convolutional Networks for Biomedical image segmentation

1) UNET의 구조 기본구조

unet은 인코더 - 디코더 기반 모델이다. 

인코딩 단계에서 입력 이미지의 특징을 포착할 수 있도록 채널의 수를 늘리면서 차원을 축소해 나간다.(!!!), 

디코딩 단계에서는 저차원으로 인코딩된 정보만을 이용하여 채널의 수를 줄이고 차원을 늘려서 고차원 이미지를 복원한다. (!!!)

- 문제점: 인코딩 단계에서 차원을 축소하면서 이미지 객체에 대한 자세한 (위치!! 정보)를 잃게 되고 디코딩 간계에서도 저차원의 정보만을 이용하기 때문에 위치정보 손실을 회복하지 못한다.

#### U-net의 기본 아이디어는 저차원 고차원 정보를 이용하여 [이미지의 특징]을 추출함과 동시에 정확한 위치 파악도 가능

- 인코딩 단계의 각 레이어의 특징을 디코딩 단계에 Concat, 인코더 레이어와 디코더 레이어의 직접연결을 스킵연걸 (skip connection)
